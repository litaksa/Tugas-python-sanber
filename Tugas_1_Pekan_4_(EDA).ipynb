##                    **Dengan Handling missing value**
![dengan handling missing value](https://drive.google.com/uc?id=1nqZuPdBsqSsOud-_XuBGujDAcJWEEvIr)

```python
# code here
df = pd.read_csv('austin_weather.csv', na_values=['na','-'])
d = {
    'DewPointAvgF' : 'float',
    'HumidityAvgPercent' : 'float',
    'WindAvgMPH' : 'float'
}
df.astype(d).info()
df.sample(1000)

#tanpa handling missing values
plt.figure(figsize=(10, 8))
plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], c=df['TempAvgF'], s=df['WindAvgMPH'].values*20, cmap='coolwarm')
plt.title('Austin Weather', fontsize=12)
plt.xlabel('Humidity Average (%)', fontsize=10)
plt.ylabel('Dew Point Average (F)', fontsize=10)
plt.xticks(fontsize=8)

#dengan handling missing values
df1 = df.fillna(method='ffill')

plt.figure(figsize=(10, 8))
plt.scatter(df1['HumidityAvgPercent'], df1['DewPointAvgF'], c=df1['TempAvgF'], s=df1['WindAvgMPH'].values*20, cmap='coolwarm')
plt.title('Austin Weather', fontsize=12)
plt.xlabel('Humidity Average (%)', fontsize=10)
plt.ylabel('Dew Point Average (F)', fontsize=10)
plt.xticks(fontsize=8)

plt.show()
```

<h2>Soal 3. Visualisasi Data with Handling outliers </h2>

Pada tugas kali ini, kita akan melakukan handling outliers. Outliers ditemukan pada kolom 'House_Price' dengan menggunakan boxplot dari library seaborn seperti gambar berikut: 


![boxplot1](https://drive.google.com/uc?id=1y2jRwf8zRPXQtSBkmjJLcqUm4l6DR7d9)


Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah
*   Visualisasikan hasil data yang sudah d remove outliers seperti expected berikut:







**EXPECTED OUTPUT:**

Batas bawah & batas atas:

871625.0 10976625.0

Visualisasi boxplot:

![boxplot2](https://drive.google.com/uc?id=1qZc-lEa5FUzSQ_cyt4_GJQ0TZ8c6WjHX)





```python
#Code here
import seaborn as sns

q1 = np.percentile(df2['House_Price'].values, 25)
q3 = np.percentile(df2['House_Price'].values, 75)
qr = q3 - q1
lb = q1 - (qr * 1.5)
ub = q3 + (qr * 1.5)

x = df2[(df2['House_Price']>lb) & (df2['House_Price']<ub)]

print(f'Batas Bawah {lb} Batas Atas {ub}')
fig, ax = plt.subplots(1,2, figsize=(18, 6))
sns.boxplot(data=df2, x='House_Price', ax=ax[0])
sns.boxplot(data=x, x='House_Price', ax=ax[1])
ax[0].set_title('Data dengan Outliers')
ax[1].set_title('Data tanpa Outliers')
plt.show()
```
