```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

#code here
sns.scatterplot(x='carat', y='price', hue='clarity', size='depth', data=df, palette=palette, sizes=(0.8, 8), linewidth=0)
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
with sns.axes_style(style='ticks'):
    sns.catplot(x='clarity', y='price', data=df, col='cut', kind='box', showfliers=False)
```

Expected Output:

![alt text](https://drive.google.com/uc?id=17mSiL-qQtFB2D8rUYB0v7d6tOrOHDoh6)

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df_new = df.sample(n=300, random_state=123).drop(['x', 'y', 'z'], axis=1)

#code here
sns.pairplot(df_new, hue='cut', kind='scatter', markers=['p', 'o', 's', 'D', 'X'], corner=True)
plt.show()
```

Expected Output:

![alt text](https://drive.google.com/uc?id=12rStINwUNGpqZd1iIGaqzTIvw9irAkJ0)

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
with sns.axes_style('white'):
    sns.jointplot(x='carat', y='price', data=df_new, kind='kde')

sns.jointplot(x='carat', y='price', data=df_new, kind='reg')
plt.show()
```
